## 1. Problema de Regresión lineal:

### 1.1. Generación de los datos:

Mediante la función `make_regressión` de sklearn, consultar su como emplearla para generar un conjunto de 2000 datos, 10 características, un ruido de 1 y fijar el parámetro ramdom_state=42.

````python
# Importación de librerías
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, make_scorer
from sklearn.model_selection import GridSearchCV
import pandas as pd
````

````python
# Genero el conjunto de datos

A , b = make_regression(
    n_samples=2000, # Número de datos / filas
    n_features=10, # Número de atributos / columnas
    noise=1,
    random_state=42 # semilla de aleatoriedad
)

print(f'Matriz A:')
display(A)
print()
print(f'Vector b:')
display (b)
````

### 1.2. Partición de datos externa.
Realizar una partición externa de tipo hold-out seleccionando un 20% de los datos para test (fijar una semilla en 42).

````python

# Hago la partición externa hold-out (80% entrenamiento, 20% test)
A_train, A_test, b_train, b_test = train_test_split(
    A, b,
    test_size=0.20, # Se crea el 20% de datos para el test
    random_state=42 # Se fija de nuevo la semilla
)

print(f"Tamaño A_train:")
display(A_train.shape)
print(f"Tamaño de A_test:")
display(A_test.shape)
````

### 1.3. Estandarización de los datos de train y test.
Utilizar el método StandardScaler().

````python
escalado = StandardScaler()

# Pasamos a ajustar y transformar los datos de entrenamiento
A_train_std = escalado.fit_transform(A_train)

# Transformo los datos de testeo
A_test_std = escalado.transform(A_test)

# Hacemos el ajuste solo en los datos de entrenamiento,
# de lo contrario, si lo hiciéramos con los datos de testeo
# estaríamos haciendo "fuga de datos"

print(f'Datos estandarizados de entrenamiento:\n')
display(A_train_std)
print()
print(f'Datos estandarizados de testeo:\n')
display(A_test_std)
````

### 1.4. Creación de modelos e hiperparametrización mediante el método de sklearn GridSearchCV:

Instrucciones:

- Cree los siguientes modelos de regresión en un diccionario:
  - Regresión Lineal (OLS).
  - Regresión Lasso
  - Regresión Rígida
  - Vecinos más cercanos (KNN)

- Cree las siguiente métricas en un diccionario para evaluar:
  - MAE
  - MSE
  - RMSE
  - R2

- Haga una búsqueda de los siguientes hiperparámetros mediante GridSearchCV:

```
# Hiperparametros
parameters = {'OLS':{},
              'Lasso':{'alpha':(0.1,0.5,1,5,10,50,100)},
              'Ridge':{'alpha':(0.1,0.5,1,5,10,50,100)},
              'KNN':{'n_neighbors':np.arange(1,15),
                     'weights':('uniform','distance'),
                     'metric':('euclidean','manhattan','cosine')
                     }
              }
```

````python
# Diccionario de modelos
modelos = {
    'Regresión Lineal (OLS)': LinearRegression(),
    'Regresión Lasso': Lasso(),
    'Regresión Rígida': Ridge(),
    'Vecinos más cercanos (KNN)': KNeighborsRegressor()
}

# Diccionario de métricas
metricas = {
    'MAE': 'neg_mean_absolute_error',
    'MSE': 'neg_mean_squared_error',
    'RMSE': 'neg_root_mean_squared_error',
    'R2': 'r2'
}

# Hiperparámetros

parametros = {
    'Regresión Lineal (OLS)': {},
    'Regresión Lasso': {'alpha': (0.1, 0.5, 1, 5, 10, 50, 100)},
    'Regresión Rígida': {'alpha': (0.1, 0.5, 1, 5, 10, 50, 100)},
    'Vecinos más cercanos (KNN)': {
        'n_neighbors': np.arange(1, 15),
        'weights': ('uniform', 'distance'),
        'metric': ('euclidean', 'manhattan', 'cosine')
    }
}
````

````python
# Se crea un diccionario para almacenar los resultados finales
resultados_grid = {}

# Itero cada modelo del diccionario que creamos antes
for nombre in modelos:
    print(f"Búsqueda iniciada para: {nombre}...")
    
    # Creanis el buscador usando GridSearchCV
    grid_search = GridSearchCV(
        estimator=modelos[nombre], # Elegimos el modelo que creamos antes
        param_grid=parametros[nombre], # Elegimos los hiperparámetros de antes.
        scoring=metricas, # Usando el diccionario de métricas creado antes.
        refit='MAE', # Paara buscar el mejor basado en el MAE
        cv=5, # Validación cruzada de 5 "folds"
        n_jobs=-1 # Usar todos los núcleos para ir más rápido
    )
    
    # Entrenamos con los datos de entrenamiento estandarizados
    grid_search.fit(A_train_std, b_train)
    
    # Guardamos el objeto entrenado en el diccionario de resultados
    resultados_grid[nombre] = grid_search
    
    # Mostramos un pequeño resumen por pantalla
    print(f"---Mejores parámetros obtenidos: {grid_search.best_params_}")
    print(f"---Mejor MAE: {-grid_search.best_score_:.4f}")
    print("-" * 60)

print("La búsqueda ha sido finalizada")
````

### 1.5. Evaluación de los modelos sobre el conjunto de test.
- Entrenar los modelos anteriores utilizando todos los datos de entrenamiento y quédese con el mejor modelo de cada uno de los modelos de regresión arrojados por GridSearchCV. Es decir, debe tener un mejor modelo por cada regresión (OLS, Lasso, Rígida y KNN)
- Evaluar su rendimiento sobre el conjunto de test mostrando para cada uno de los modelos

````python
# Creamos otro diccionario para almacenar las métricas finales del testeo
resumen_test = []

for nombre in resultados_grid:
    # Extraemos el mejor estimador del modelo ya entrenado con los mejores parámetros
    mejor_modelo = resultados_grid[nombre].best_estimator_
    
    # Hacemos la predicción sobre el conjunto de testeo
    b_pred = mejor_modelo.predict(A_test_std)
    
    # Calculamos las métricas de test
    mae = mean_absolute_error(b_test, b_pred)
    mse = mean_squared_error(b_test, b_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(b_test, b_pred)
    
    # Almaceno los resultados en una lista
    resumen_test.append({
        'Modelo': nombre,
        'MAE Test': mae,
        'MSE Test': mse,
        'RMSE Test': rmse,
        'R2 Test': r2
    })

# Convertimos a DataFrame para mejorar la visualzación
df_resultados = pd.DataFrame(resumen_test)
display(df_resultados)
````

### 1.6. Interpretación de resultados.
* Justifica cuál de los modelos utilizarías para ponerlo en producción

De acuerdo con los resultados que se pueden ver, podríamos decir que la mejor elección sería usar la **regresión rígida** y en segunda posición la **regresión lineal tradicional**

- Mejores métricas en Test: Es el modelo con el error más bajo de todos con un MAE de 0.7726 y un RMSE de 0.9759. Supera ligeramente a la regresión lineal tradicional (OLS).
- Ajuste casi perfecto: Tiene un $R^2$ de 0.9999, lo que significa que el modelo explica prácticamente toda la varianza de los datos.
- Robustez: Al usar penalización $L2$ (con $\alpha=0.1$), el modelo parece ser más estable frente al ruido que el OLS normal. Esto nos ayuda a evitar un posible sobreajuste.
- KNN descartado: El modelo KNN no funciona bien en este caso (MAE de 50.04).

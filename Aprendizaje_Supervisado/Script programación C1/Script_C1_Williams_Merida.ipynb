## 1. Problema de Regresión lineal:

### 1.1. Generación de los datos:

Mediante la función `make_regressión` de sklearn, consultar su como emplearla para generar un conjunto de 2000 datos, 10 características, un ruido de 1 y fijar el parámetro ramdom_state=42.

````python
# Importación de librerías
import numpy as np
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LinearRegression, Lasso, Ridge
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, make_scorer
from sklearn.model_selection import GridSearchCV
import pandas as pd
````

````python
# Genero el conjunto de datos

A , b = make_regression(
    n_samples=2000, # Número de datos / filas
    n_features=10, # Número de atributos / columnas
    noise=1,
    random_state=42 # semilla de aleatoriedad
)

print(f'Matriz A:')
display(A)
print()
print(f'Vector b:')
display (b)
````

### 1.2. Partición de datos externa.
Realizar una partición externa de tipo hold-out seleccionando un 20% de los datos para test (fijar una semilla en 42).

````python

# Hago la partición externa hold-out (80% entrenamiento, 20% test)
A_train, A_test, b_train, b_test = train_test_split(
    A, b,
    test_size=0.20, # Se crea el 20% de datos para el test
    random_state=42 # Se fija de nuevo la semilla
)

print(f"Tamaño A_train:")
display(A_train.shape)
print(f"Tamaño de A_test:")
display(A_test.shape)
````

### 1.3. Estandarización de los datos de train y test.
Utilizar el método StandardScaler().

````python
escalado = StandardScaler()

# Pasamos a ajustar y transformar los datos de entrenamiento
A_train_std = escalado.fit_transform(A_train)

# Transformo los datos de testeo
A_test_std = escalado.transform(A_test)

# Hacemos el ajuste solo en los datos de entrenamiento,
# de lo contrario, si lo hiciéramos con los datos de testeo
# estaríamos haciendo "fuga de datos"

print(f'Datos estandarizados de entrenamiento:\n')
display(A_train_std)
print()
print(f'Datos estandarizados de testeo:\n')
display(A_test_std)
````

### 1.4. Creación de modelos e hiperparametrización mediante el método de sklearn GridSearchCV:

Instrucciones:

- Cree los siguientes modelos de regresión en un diccionario:
  - Regresión Lineal (OLS).
  - Regresión Lasso
  - Regresión Rígida
  - Vecinos más cercanos (KNN)

- Cree las siguiente métricas en un diccionario para evaluar:
  - MAE
  - MSE
  - RMSE
  - R2

- Haga una búsqueda de los siguientes hiperparámetros mediante GridSearchCV:

```
# Hiperparametros
parameters = {'OLS':{},
              'Lasso':{'alpha':(0.1,0.5,1,5,10,50,100)},
              'Ridge':{'alpha':(0.1,0.5,1,5,10,50,100)},
              'KNN':{'n_neighbors':np.arange(1,15),
                     'weights':('uniform','distance'),
                     'metric':('euclidean','manhattan','cosine')
                     }
              }
```

````python
# Diccionario de modelos
modelos = {
    'Regresión Lineal (OLS)': LinearRegression(),
    'Regresión Lasso': Lasso(),
    'Regresión Rígida': Ridge(),
    'Vecinos más cercanos (KNN)': KNeighborsRegressor()
}

# Diccionario de métricas
metricas = {
    'MAE': 'neg_mean_absolute_error',
    'MSE': 'neg_mean_squared_error',
    'RMSE': 'neg_root_mean_squared_error',
    'R2': 'r2'
}

# Hiperparámetros

parametros = {
    'Regresión Lineal (OLS)': {},
    'Regresión Lasso': {'alpha': (0.1, 0.5, 1, 5, 10, 50, 100)},
    'Regresión Rígida': {'alpha': (0.1, 0.5, 1, 5, 10, 50, 100)},
    'Vecinos más cercanos (KNN)': {
        'n_neighbors': np.arange(1, 15),
        'weights': ('uniform', 'distance'),
        'metric': ('euclidean', 'manhattan', 'cosine')
    }
}
````

````python
# Se crea un diccionario para almacenar los resultados finales
resultados_grid = {}

# Itero cada modelo del diccionario que creamos antes
for nombre in modelos:
    print(f"Búsqueda iniciada para: {nombre}...")
    
    # Creanis el buscador usando GridSearchCV
    grid_search = GridSearchCV(
        estimator=modelos[nombre], # Elegimos el modelo que creamos antes
        param_grid=parametros[nombre], # Elegimos los hiperparámetros de antes.
        scoring=metricas, # Usando el diccionario de métricas creado antes.
        refit='MAE', # Paara buscar el mejor basado en el MAE
        cv=5, # Validación cruzada de 5 "folds"
        n_jobs=-1 # Usar todos los núcleos para ir más rápido
    )
    
    # Entrenamos con los datos de entrenamiento estandarizados
    grid_search.fit(A_train_std, b_train)
    
    # Guardamos el objeto entrenado en el diccionario de resultados
    resultados_grid[nombre] = grid_search
    
    # Mostramos un pequeño resumen por pantalla
    print(f"---Mejores parámetros obtenidos: {grid_search.best_params_}")
    print(f"---Mejor MAE: {-grid_search.best_score_:.4f}")
    print("------------------------------------------------")

print("La búsqueda ha sido finalizada")
````

### 1.5. Evaluación de los modelos sobre el conjunto de test.
- Entrenar los modelos anteriores utilizando todos los datos de entrenamiento y quédese con el mejor modelo de cada uno de los modelos de regresión arrojados por GridSearchCV. Es decir, debe tener un mejor modelo por cada regresión (OLS, Lasso, Rígida y KNN)
- Evaluar su rendimiento sobre el conjunto de test mostrando para cada uno de los modelos

````python
# Creamos otro diccionario para almacenar las métricas finales del testeo
resumen_test = []

for nombre in resultados_grid:
    # Extraemos el mejor estimador del modelo ya entrenado con los mejores parámetros
    mejor_modelo = resultados_grid[nombre].best_estimator_
    
    # Hacemos la predicción sobre el conjunto de testeo
    b_pred = mejor_modelo.predict(A_test_std)
    
    # Calculamos las métricas de test
    mae = mean_absolute_error(b_test, b_pred)
    mse = mean_squared_error(b_test, b_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(b_test, b_pred)
    
    # Almaceno los resultados en una lista
    resumen_test.append({
        'Modelo': nombre,
        'MAE Test': mae,
        'MSE Test': mse,
        'RMSE Test': rmse,
        'R2 Test': r2
    })

# Convertimos a DataFrame para mejorar la visualzación
df_resultados = pd.DataFrame(resumen_test)
display(df_resultados)
````

### 1.6. Interpretación de resultados.
* Justifica cuál de los modelos utilizarías para ponerlo en producción

De acuerdo con los resultados que se pueden ver, podríamos decir que la mejor elección sería usar la **regresión rígida** y en segunda posición la **regresión lineal tradicional**

- Mejores métricas en Test: Es el modelo con el error más bajo de todos con un MAE de 0.7726 y un RMSE de 0.9759. Supera ligeramente a la regresión lineal tradicional (OLS).
- Ajuste casi perfecto: Tiene un $R^2$ de 0.9999, lo que significa que el modelo explica prácticamente toda la varianza de los datos.
- Robustez: Al usar penalización $L2$ (con $\alpha=0.1$), el modelo parece ser más estable frente al ruido que el OLS normal. Esto nos ayuda a evitar un posible sobreajuste.
- KNN descartado: El modelo KNN no funciona bien en este caso (MAE de 50.04).

## 2. Problema de Clasificación con generación de datos sintéticos y selección de características

En esta actividad se trabajará con generación de datos sintéticos para problemas de clasificación utilizando la función `make_classification` de la librería **scikit-learn**.

El objetivo es evaluar si distintos métodos de selección de características y modelos de clasificación son capaces de identificar las variables realmente relevantes dentro de un conjunto de datos generado artificialmente.

### 2.1 Importación de librerías

Importe las siguientes librerías:

````python
import numpy as np
import pandas as pd

from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.metrics import accuracy_score, f1_score

from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif

from sklearn.preprocessing import StandardScaler
````

### 2.2. Generación del conjunto de datos

2.2.1. Utilice `make_classification` para generar un dataset de clasificación binaria.

Debe definir explícitamente los siguientes parámetros:

- `n_samples`
- `n_features`
- `n_informative`
- `n_redundant`

El dataset debe tener:

- al menos **500 observaciones**
- al menos **10 características**

El objetivo es que existan variables:

- **informativas**
- **redundantes**
- **no informativas**

2.2.2. Explique brevemente los valores elegidos para cada parámetro.


````python
# Generamos el dataset de clasificación
A, b = make_classification(
    n_samples=1000,       
    n_features=20,        
    n_informative=5,      
    n_redundant=3,        
    n_clusters_per_class=1,
    random_state=42       
)

print(f"Tamaño de A:")
display(A.shape)
print(f"Tamaño de b:")
display(b.shape)
````

He seleccionado estos parámetros aplicando un criterio de robustez similar al coeficiente de seguridad en ingeniería:

- **n_samples=1000**: He duplicado el mínimo requerido (500) para asegurar una estadística más estable y reducir la incertidumbre en los resultados.

- **n_features=20**: Duplicando el valor de las características para un entorno de prueba más exigente. Al tener 12 variables de ruido (el 60% del dataset), el proceso de selección de características será más crítico para evitar el sobreajuste.

- **n_informative=5** y **n_redundant=3**: Mantengo un núcleo controlado de variables con señal (8 en total), dejando que el resto del dataset sea "paja" que el algoritmo debe aprender a descartar.

- **random_state=42**: He decidido dejar la semilla como en los apartados anteriores

### 2.3. División de los datos.

Divida el dataset en:

- **conjunto de entrenamiento**
- **conjunto de prueba**

El conjunto de entrenamiento se utilizará para:

- selección de características
- búsqueda de hiperparámetros
- entrenamiento de los modelos

El conjunto de prueba se utilizará **únicamente para evaluar el desempeño final de los modelos**.

````python
# Divido el dataset en 80% entrenamiento y 20% prueba.
A_train, A_test, b_train, b_test = train_test_split(
    A, b, 
    test_size=0.20, 
    random_state=42, 
    stratify=b # Para una misma proporción de clases en ambos grupos
)

print(f"Muestras de entrenamiento:")
display(A_train.shape[0])
print()
print(f"Muestras de prueba:")
display(A_test.shape[0])
````

### 2.4. Selección de caractarística

Aplique un método de selección de características para identificar cuáles variables parecen ser más relevantes para la predicción.

Puede utilizar el método **SelectKBest** de `sklearn.feature_selection`. Este método evalúa cada característica de forma individual mediante una medida estadística y selecciona las **k variables con mayor puntuación**, es decir, aquellas que tienen mayor relación con la variable objetivo.

Utilice como función de puntuación `f_classif`.

Compare las características seleccionadas con las definidas como **informativas** en `make_classification` y discuta si el método logra identificarlas correctamente.

````python
# Definimos que busque las k=5 mejores 
# Usamos solo los datos de entrenamiento para no "contaminar" el proceso
selector = SelectKBest(score_func=f_classif, k=5)
selector.fit(A_train, b_train)

# extreamos los índices de las variables seleccionadas
indices_seleccionados = selector.get_support(indices=True)

# Mostramos las puntuaciones de todas las variables para analizar el "ruido"
puntuaciones = selector.scores_

print(f'Índices de las 5 variables elegidas:')
print(indices_seleccionados)

# Creamos una comparativa rápida
df_scores = pd.DataFrame({'Característica / Feature': range(20), 'Puntuación': puntuaciones})
df_scores = df_scores.sort_values(by='Puntuación', ascending=False)
display(df_scores.head(10)) # Mostramos las 10 mejores para ver el corte
````

## 2.5. Escalado de datos

Antes de entrenar los modelos, estandarice las variables utilizando **StandardScaler** de `sklearn.preprocessing`.

La estandarización consiste en transformar las variables para que tengan:

- media igual a 0
- desviación estándar igual a 1

Este paso es importante para modelos sensibles a la escala de los datos, como **regresión logística** y **SVM**. Los árboles de decisiones, son independientes del escalado.

El escalado debe realizarse de la siguiente manera:

1. Ajustar el `StandardScaler` **únicamente con el conjunto de entrenamiento**.
2. Aplicar esa transformación tanto al conjunto de entrenamiento como al conjunto de prueba.


````python
# Creamos el objeto como hicimos antes
escalado2 = StandardScaler()

# Ajustamos para aprender la media y desviación típica de A_train
escalado2.fit(A_train)

# Transformamos ambos conjuntos
A_train_std = escalado2.transform(A_train)
A_test_std = escalado2.transform(A_test)

A_train_std_media = round(A_train_std.mean(), 2)
A_train_std_desv = round(A_train_std.std(), 2)

print(f"Media de A_train_std (debe ser aprox. 0):")
display(A_train_std_media)
print()
print(f"Desviación de A_train_std (debe ser aprox. 1):")
display(A_train_std_desv)
````

### 2.6. Modelos e hiperparametrización

Trabaje con los siguientes modelos.

**Modelos obligatorios**

- Árbol de decisión (`DecisionTreeClassifier`)
- Regresión logística (`LogisticRegression`)

**Modelo bonus**

- Máquina de soporte vectorial (`SVC`)

Para cada modelo, realice una búsqueda de hiperparámetros utilizando **GridSearchCV** sobre el **conjunto de entrenamiento**.

Utilice las siguientes listas de hiperparámetros:

**Árbol de decisión**

- `max_depth`: [3, 5, 10, None]
- `min_samples_split`: [2, 5, 10]
- `criterion`: ["gini", "entropy"]

**Regresión logística**

- `C`: [0.01, 0.1, 1, 10]
- `solver`: ["lbfgs", "liblinear"]

**SVM (bonus)**

- `C`: [0.1, 1, 10]
- `kernel`: ["linear", "rbf"]
- `gamma`: ["scale", "auto"]

Utilice validación cruzada dentro de `GridSearchCV` para determinar la mejor combinación de hiperparámetros.


````python
# Diccionario de modelos
modelos_clas = {
    'Árbol de Decisión': DecisionTreeClassifier(random_state=42),
    'Regresión Logística': LogisticRegression(max_iter=1000, random_state=42),
    'SVM (Bonus)': SVC(probability=True, random_state=42)
}

# Diccionario de hiperparámetros
parametros_clas = {
    'Árbol de Decisión': {
        'max_depth': [3, 5, 10, None],
        'min_samples_split': [2, 5, 10],
        'criterion': ["gini", "entropy"]
    },
    'Regresión Logística': {
        'C': [0.01, 0.1, 1, 10],
        'solver': ["lbfgs", "liblinear"]
    },
    'SVM (Bonus)': {
        'C': [0.1, 1, 10],
        'kernel': ["linear", "rbf"],
        'gamma': ["scale", "auto"]
    }
}
````

````python
resultados_clas = {}

for nombre in modelos_clas:
    print(f"Entrenando el {nombre}...")
    
    grid = GridSearchCV(
        estimator=modelos_clas[nombre],
        param_grid=parametros_clas[nombre],
        cv=5,
        scoring='accuracy', 
        n_jobs=-1
    )
    
    # Usamos los datos que estandarizamos antes
    grid.fit(A_train_std, b_train)
    
    resultados_clas[nombre] = grid
    
    print(f"---Mejores parámetros obtenidos: {grid.best_params_}")
    print(f"---Mejor precisión obtenida: {grid.best_score_:.4f}")
    print("------------------------------------------------")

print("Proceso finalizado.")
````

### .7. Evaluación y análisis

Para cada modelo:

- Del punto anterior, obtenga el modelo con **los mejores hiperparámetros**.
- Evalúe el modelo final en el **conjunto de prueba**.
- Realice este proceso en dos escenarios:
  - utilizando **todas las características**
  - utilizando **solo las características seleccionadas**
- Compare los resultados utilizando métricas de clasificación apropiada y discuta:
  - si la selección de características mejora el desempeño del modelo
  - si reduce la complejidad del modelo
  - si permite identificar correctamente las variables informativas generadas en el dataset
  - qué modelo obtiene mejor desempeño después del ajuste de hiperparámetros


````python
# Caso 1: usando todas características (las 20)
# Caso 2: Usar solo las seleccionadas por SelectKBest. Las k=5 mejores

resultados_comparativa = []

for nombre in resultados_clas:
    # Modelo con el caso 1
    mejor_modelo_all = resultados_clas[nombre].best_estimator_
    pred_all = mejor_modelo_all.predict(A_test_std)
    acc_all = accuracy_score(b_test, pred_all)
    
    # Modelo con el caso 2
    # Se filtra primero los datos de entrenamiento y test para la 5 columnas.
    A_train_sel = A_train_std[:, indices_seleccionados]
    A_test_sel = A_test_std[:, indices_seleccionados]
    
    # Volvemos a entrenar el modelos mejorado pero solo con las 5 variables
    mejor_modelo_sel = mejor_modelo_all.fit(A_train_sel, b_train)
    pred_sel = mejor_modelo_sel.predict(A_test_sel)
    acc_sel = accuracy_score(b_test, pred_sel)
    
    resultados_comparativa.append({
        'Modelo': nombre,
        'Accuracy (Todas)': acc_all,
        'Accuracy (Seleccionadas)': acc_sel,
        'Diferencia': acc_sel - acc_all
    })

df_final = pd.DataFrame(resultados_comparativa)
display(df_final)
````

La selección de variables ha funcionado bien, especialmente en la Regresión Logística, donde la precisión subió de 0.960 a 0.975. En SVM y Árbol la bajada es mínima, lo que confirma que las 12 variables de ruido no nos aportaban nada en especial.

- Complejidad y Variables Informativas:

    - Hemos reducido el dataset en un 75% (de 20 a 5 variables) manteniendo un rendimiento casi idéntico. Esto hace al modelo mucho más ligero y fácil de interpretar.

    - Identificación: El método SelectKBest ha funcionado correctamente al filtrar el ruido y quedarse con el núcleo informativo que definimos en la generación.

- **Mejor modelo**:
El mejor desempeño lo obtiene la Regresión Logística con selección de variables (0.975). Ha resultado ser el modelo más robusto para generalizar sobre los datos de prueba una vez eliminado el ruido.
